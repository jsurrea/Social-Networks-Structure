```python
import networkx as nx
import pandas as pd

# Cargar los datos de la red de amistad desde un archivo CSV
data_file = "Data Set.xlsx"  # Nombre del archivo que contiene los datos de la red
sheet_name = "t=2"  # Nombre de la pestaña que contiene los datos en el archivo

# Cargar los datos desde la pestaña en un DataFrame de pandas
data = pd.read_excel(data_file, sheet_name=sheet_name)

# Crear un grafo dirigido a partir de los datos
G_t2 = nx.Graph()

for i in range(1,51):
    G_t2.add_node(i)

# Agregar nodos y aristas al grafo a partir de los datos
for index, row in data.iterrows():
    source_node = row["Source"]
    target_node = row["Target"]
    weight = row["Weight"]
    G_t2.add_node(source_node)
    G_t2.add_node(target_node)
    G_t2.add_edge(source_node, target_node, weight=weight)

# Ahora tienes la red de amistad cargada en el grafo G_t2
```

```python
import networkx as nx

# Supongamos que tienes un grafo G_t1 creado a partir de tus datos.

# Encontrar componentes conectados
connected_components = list(nx.connected_components(G_t1))

# Calcular la eficiencia de red para cada componente conectado
for i, component in enumerate(connected_components):
    subgraph = G_t1.subgraph(component)
    try:
        average_shortest_path_length = nx.average_shortest_path_length(subgraph)
        efficiency = 1 / average_shortest_path_length
        print(f"Componente Conectado {i + 1}: Eficiencia de Red = {efficiency}")
    except nx.NetworkXError:
        print(f"Componente Conectado {i + 1}: No es posible calcular la eficiencia (componente desconectado)")

```

```python
import networkx as nx

# Supongamos que tienes dos grafos G_t1 y G_t2 para los momentos de tiempo T=1 y T=2 respectivamente.

# Calcular el promedio de grados en T=1
avg_degree_t1 = sum(dict(G_t1.degree()).values()) / len(G_t1)

# Calcular el promedio de grados en T=2
avg_degree_t2 = sum(dict(G_t2.degree()).values()) / len(G_t2)

# Comparar los valores y determinar si la cantidad de conexiones aumenta o disminuye
if avg_degree_t1 < avg_degree_t2:
    print("La cantidad de conexiones entre estudiantes aumenta con el tiempo.")
elif avg_degree_t1 > avg_degree_t2:
    print("La cantidad de conexiones entre estudiantes disminuye con el tiempo.")
else:
    print("La cantidad de conexiones entre estudiantes se mantiene constante con el tiempo.")

    
print(f"T=1 Promedio de grados: {avg_degree_t1}")
print(f"T=2 Promedio de grados: {avg_degree_t2}")
```

```python
import pandas as pd

# Combinar los DataFrames relevantes en attributes_data_t2
attributes_data_t2 = pd.DataFrame({"Node": alcohol_data["Node"]})

# Agregar las columnas de las variables de interés de t=2
attributes_data_t2["Alcohol"] = alcohol_data["t=2"]
attributes_data_t2["Drugs"] = drugs_data["t=2"]
attributes_data_t2["Smoke"] = smoke_data["t=2"]
attributes_data_t2["Sport"] = sport_data["t=2"]

attributes_data_t2.set_index("Node", inplace=True)

# Mostrar los primeros 5 registros del DataFrame attributes_data_t2
print("Primeros 5 registros de attributes_data_t2:")
print(attributes_data_t2.head(5))
```

```python
# Supongamos que tienes un DataFrame llamado attributes_data_t2 con una columna "Drugs" para t=2.

def transform_drugs(row):
    if row["Drugs"] in [1, 2]:
        return "Baja"
    else:
        return "Alta"

predisposition = attributes_data_t2.apply(transform_drugs, axis=1)
```
